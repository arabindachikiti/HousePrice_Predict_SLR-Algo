```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
HousePrice_dataset= pd.read_csv(r'F:\Data science\April Month\21-04-2025\21st- SLR\SLR - House price prediction\House_data.csv')
```

```python
HousePrice_dataset.head()
```

```python
space=HousePrice_dataset['sqft_living'].values
price=HousePrice_dataset['price'].values
```

```python
x=np.array(space).reshape(-1,1)
y=np.array(price).reshape(-1,1)
print(x.shape)
print(y.shape)
```

Spliting the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

Filling Simple Linear Regression to the Traing set

```python
pip install sklearn.linear_model
```

```python

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

#Predicting the prices


```python
predict=regressor.predict(x_test)
```

Visualization traing set result

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('House Price Prediction')
plt.xlabel('space')
plt.ylabel('price')
plt.show()
```

Visualization of Test result

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,regressor.predict(x_train) ,color='blue')
plt.title('House Price Prediction')
plt.xlabel('space') 
plt.ylabel('price')
plt.show()
```

Slope and Coficient

```python
m_slope=regressor.coef_
print(m_slope)
c_intercept=regressor.intercept_
print(c_intercept)
```

House price predition


```python
house_price_predition=m_slope*1100+c_intercept
print(house_price_predition)
```

```python
#Bias
bias=regressor.score(x_train,y_train)
print(bias)
```

```python
#Variance
variance=regressor.score(x_test,y_test)
print(variance)
```
